# Map, Apply, and Transform

```python
# memetakan value tertentu menjadi value tertentu
df['shelf_baru'] = df.shelf.map({
    1: "bawah",
    2: "tengah",
    3: "atas"
})
df.head()
```

```python
# apply -> melakukan sesuatu pada seluruh kolom itu
df['fat_mg'] = df.fat.apply(lambda x: x * 1000)
df.head()
```

```python
# Scalling
# penskalaan dari skala lama ke 0 sampai 1
# misal nilai min adalah 3 dan maksimal 13 diskalakan baru menjadi 0 sampai 1
# rumus:
# x_new = (x-x_min) / (x_max-x_min)
```

```python
# beda transform dan apply:
# transform -> x yang dimasukkan adalah semua x yang ada di df, jadi ada x.min nya dan x.maxnya dari semua x di df
# apply -> x yang dimasukkan adalah x pada saat itu aja, jadi x.min nya dan x.maxnya yang x itu sendiri

df['scaled_rating'] = df.rating.transform( lambda x: (x-x.min())/(x.max()-x.min()) )
df
```

```python
df.drop(columns=["shelf_baru", "fat_mg", "scaled_rating"], inplace=True)
```

# Konsep Encoding pada Data Kategorikal
- Untuk data yang bersifat kategorikal(misal: mfr yang berisi manufacturer2 biskuit)
- machine learning tidak bisa menerima yang tipenya object atau bahkan string. Oleh karena itu data kategorikal perlu di encode menjadi angka (numerikal)
- Ada 2 cara:
     - label encoding (ordinal encoding)
     - dummy variable encoding (one hot encoding)
- label encoding: 
     -tiap kategori direpresentasikan dalam angka
     - misal: 
         - bawah -> 0
         - tengah -> 1
         - atas -> 2
- one hot encoding (lebih banyak dipakai karena lebih fleksibel):          
    | shelfnya | isBawah | isTengah | isAtas |
    |----------|---------|----------|--------|
    | bawah    | 1       | 0        | 0      |
    | tengah   | 0       | 1        | 0      |
    | atas     | 0       | 0        | 1      |         
  

# One Hot Encoding / Dummy Variable

# Binning (Kotak2in)
- Kebalikan dari sebelumnya -> mengubah data numerikal menjadi kategorikal
- misal umur
|Usia    |Kategori   |
|--------|-----------|
| 0-5    | balita    |
| 5-10   | anak-anak |
| 10-18  | remaja    |
| 18-40  | dewasa    |
| 40-120 | lansia    |

```python
# Mengkategorikan rating menjadi grade A, B, dan C
# binsnya ada 4 sedangkan labelnya ada 3 karena bins itu interval
# jadi dari 0-40 -> grade C, 41-70 -> grade B, 71-maks -> grade A
df['grade'] = pd.cut(df.rating, bins=[0, 40, 70, df.rating.max()], labels=["C", "B", "A"])
df.head()
```
