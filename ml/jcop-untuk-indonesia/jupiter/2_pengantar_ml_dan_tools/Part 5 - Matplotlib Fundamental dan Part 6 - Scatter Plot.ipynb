```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0,10,100)
y = np.sqrt(X) # akarin member2nya
y1 = np.sin(X)
y2 = np.cos(X)
```

# Inisiasi Figure dan Axes (Figure(Kotaknya) dan Koordinat(X dan Y nya))

```python
fig  = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure(figsize=(7,7)) # ukuran figure
ax = plt.axes() # koordinat
ax.plot(X,y) # isi koordinat
```

```python
fig = plt.figure(figsize=(6,6)) # ukuran figure
ax = plt.axes() # koordinat
# jika isinya lebih dari 1 
ax.plot(X,y) # plot 1
ax.plot(X,y1) # plot 2
ax.plot(X,y2) # plot 3
```

```python
# cari singkat
plt.figure(figsize=(6,6))
plt.plot(X,y) 
plt.plot(X,y1)
plt.plot(X,y2)
```

```python
# styling garisnya
plt.figure(figsize=(6,6))
plt.plot(X,y1, 'r-.') # garis warna merah strip2
plt.plot(X,y2, 'b-') # garis warna biru garis lurus
```

```python
# limit sumbu x dan y
plt.figure(figsize=(6,6))
plt.plot(X,y1, 'r-.') 
plt.plot(X,y2, 'b-') 
plt.xlim(0,20) # sumbu x dari 0 sampai 20
plt.ylim(-2, 5) # sumbu y dari -2 sampai 5
```

```python
plt.figure(figsize=(6,6))
plt.plot(X,y1, 'r-', linewidth=0.5) # tebal baris
plt.plot(X,y2, 'bo', markersize=2)  # ukuran marker
```

```python
plt.figure(figsize=(6,6))
plt.plot(X,y1, 'r-') 
plt.plot(X,y2, 'bo')  
plt.axis('equal') # skala x dan y disetarakan
```

# Scatter Plot (Plot tapi pakai marker doang tanpa garis)

```python
df = pd.read_csv('matplotlib_data/circle_data.csv', delimiter=";")
df.head()
```

```python
X = df.x
y = df.y
rad = df.radius
```

```python
plt.scatter(X, y)
plt.axis('equal')
```

```python
plt.scatter(X, y, c='r', s=10) # warna merah dan size tiap marker 10
plt.axis('equal')
```

### perbedaan scatter dan plot
- scatter itu tiap nilainya bisa memiliki marker yang berbeda2 tergantung sifat datanya, sedangkan plot markernya ya cuma bisa diset 1 untuk semua

```python
plt.scatter(X, y, c='r', s=rad*40) # warna merah dan size tiap marker 10
plt.axis('equal')
```

```python
plt.scatter(X, y, c=rad, s=rad*40) # atur warna dan ukuran tiap data berdasarkan field radius
plt.axis('equal')
```
